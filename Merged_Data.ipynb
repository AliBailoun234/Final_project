```python
# Initial imports
import pandas as pd
import hvplot.pandas
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
# Load the crypto_data.csv dataset.
seasons_stats = "Files/Seasons_Stats.xlsx"
players = "Files/Players.xlsx"
player_data = "Files/player_data.xlsx"
seasons_stats_df = pd.read_excel(seasons_stats)
players_df = pd.read_excel(players)
player_data_df = pd.read_excel(player_data)
```

```python
join_df1 = seasons_stats_df.merge(players_df,on='Player',how='left')
```

```python
merged_df = join_df1.merge(player_data_df,left_on='Player',right_on='name',how='left')
```

```python
merged_df = merged_df.drop(columns=['name','position','height_y','weight_x','college_y','birth_date'])
```

```python
Decades = []
for i in range(len(merged_df)):
    val = merged_df.iloc[i,0]
    if (val>=1950) & (val<1960):
        Decades.append('1950s')
    elif (val>=1960) & (val<1970):
        Decades.append('1960s')
    elif (val>=1970) & (val<1980):
        Decades.append('1970s')
    elif (val>=1980) & (val<1990):
        Decades.append('1980s')
    elif (val>=1990) & (val<2000):
        Decades.append('1990s')
    elif (val>=2000) & (val<2010):
        Decades.append('2000s')
    elif (val>=2010) & (val<2020):
        Decades.append('2010s')
    else:
        Decades.append('N/A')
```

```python
merged_df['Decade'] = Decades
```

```python
merged_df = merged_df.loc[merged_df['Decade'] != 'N/A']
```

```python
merged_df.to_csv('Files/merged_data.csv',index=False)
```
