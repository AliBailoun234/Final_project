```python
Merge_tables = pd.concat([Merge1, player_data])
Merge_tables
```

```python
Merge_filter_pd = Merge_tables["Year"].between(1970,1985,inclusive = True)
Merge_tables[Merge_filter_pd]
```

```python
drop_pd = Merge_tables[Merge_filter_pd].drop(["collage","born","birth_city","birth_state","name","year_start","year_end","position","birth_date","college"],axis=1)
drop_pd
```

```python
drop1_pd = drop_pd.drop_duplicates(subset=["Player"])
drop1_pd
```

```python
drop2_pd = drop1_pd.drop(["Tm"],axis=1)
drop2_pd
```

```python
drop3_pd = drop2_pd.drop(["Player","Pos"],axis=1)
drop3_pd['height'] = drop3_pd['height'].astype(float)
drop4_pd = drop3_pd.dropna()
drop4_pd.columns
```

```python
import pandas as pd
import hvplot.pandas
from path import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Select the variables to use in the analysis
X = drop4_pd[['TS%', 'FTr', '2P%', 'eFG%', 'FT%','Age']]

scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(X_std)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Perform k-means clustering with the chosen number of clusters
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans
```

```python
# Fitting model
kmeans.fit(X)

# Get the predictions
predictions = kmeans.predict(X)
print(predictions)
```

```python
# Add a new class column to the df_iris
X["class"] = kmeans.labels_
X.head()
```

```python
# Create a scatterplot of df_iris
X.hvplot.scatter(x="TS%", y="FTr",by="class")
```

```python
# Plotting the clusters with three features
fig = px.scatter_3d(X, x="TS%", y="FTr", z="2P%", color="class", symbol="class", size="FT%",width=800)
fig.update_layout(legend=dict(x=0,y=1))
fig.show()
```

```python
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of K values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(X)
    inertia.append(km.inertia_)
```

```python
# Create elbow data
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k",y="inertia",xticks=k,title="Elbow Curve")
```

```python
def get_clusters(k, data):
    # Create a copy of the DataFrame
    data = data.copy()

    # Initialize the K-Means model
    model = KMeans(n_clusters=k, random_state=0)

    # Fit the model
    model.fit(data)

    # Predict clusters
    predictions = model.predict(data)

    # Create return DataFrame with predicted clusters
    data["class"] = model.labels_

    return data
```

```python
three_clusters = get_clusters(3, X)
three_clusters.head()
```

```python
# Create a scatterplot
X.hvplot.scatter(x="Age", y="FT%",by="class")
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

M = drop4_pd[['TS%', '2P%', 'eFG%', 'FT%','Age','height','weight']]

# Standardize the data
scaler = StandardScaler()
X_std = scaler.fit_transform(M)

# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(X_std)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Perform k-means clustering with the chosen number of clusters
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)
kmeans.fit(X_std)

# Add the cluster labels to the original data
drop4_pd['Cluster'] = kmeans.labels_

# Plot the results
fig, ax = plt.subplots(figsize=(8, 6))
colors = {0:'red', 1:'blue', 2:'green'}
ax.scatter(drop4_pd['weight'], drop4_pd['2P%'], c=drop4_pd['Cluster'].apply(lambda x: colors[x]))
ax.set_xlabel('weight')
ax.set_ylabel('2 P %')
ax.set_title('Cluster Analysis of NBA Players')
plt.show()
```
